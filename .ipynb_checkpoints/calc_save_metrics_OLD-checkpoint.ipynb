# Calculate & Save Metrics

**Purpose**: This script calculates 3 evaluation metrics that are used to populate tables and figures for the ClimSim manuscript:
- Mean absolute error ($\text{MAE}$)
- Root mean squared error ($\text{RMSE}$)
- R squared ($R{^2}$)
 
**Output**:
Save these metrics as a ```.csv``` file with names set in ```metrics_3D_file``` and ```metrics_2D_file```.

## Import Necessary Packages

````python
import numpy as np
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
````

## Define Inputs

````python
# Model names and where their outputs are located
model_output = {'MLP' :'./model_outputs/001_backup_phase-7_retrained_models_step2_lot-147_trial_0027.best.h5.npy',
                'RPN' :'./model_outputs/rpn_pred_v1_stride6.npy',
                'CNN' :'./model_outputs/val_predict_cnn_reshaped_stride6_FINAL.npy',
                'cVAE':'./model_outputs/cvae.h5',
                'HSR' :'./model_outputs/hsr_preds_bestcrps.h5'}

# Where the grid information is located
grid_file = '../../misc/E3SM-MMF_ne4_grid-info.nc'

# The model name used for the output
model_name = 'MLP'

# Where the .npy files for validation set input, output, and model predictions are located
x_true_file = '../../npy_data_conversion/npy_files/val_input_stride6.npy'
y_true_file = '../../npy_data_conversion/npy_files/val_target_stride6.npy'
y_pred_file = model_output[model_name]

# Where the .nc files for input and output normalization/scaling coefficients are located
in_mean_file   = '../../norm_factors/mli_mean.nc'
in_min_file    = '../../norm_factors/mli_min.nc'
in_max_file    = '../../norm_factors/mli_max.nc'
out_scale_file = '../../norm_factors/mlo_scale.nc'

# Where to save the .csv files for 3D and 2D (level-averaged) metric output
metrics_3D_file = f'./metrics/{model_name}.metrics.csv'
metrics_2D_file = f'./metrics/{model_name}.metrics.lev-avg.csv'
````

## Load Data

````python
# Load in the input data
x_true = np.load(x_true_file).astype(np.float64)
y_true = np.load(y_true_file).astype(np.float64)
if y_pred_file[-3:] == '.h5':
    y_pred = xr.open_dataset(y_pred_file)['pred'].values
else:
    y_pred = np.load(y_pred_file).astype(np.float64)

# Load in the normalization/scaling coefficients
in_mean   = xr.open_dataset(in_mean_file)
in_min    = xr.open_dataset(in_min_file)
in_max    = xr.open_dataset(in_max_file)
out_scale = xr.open_dataset(out_scale_file)

# Load the grid information (ncol = 384, where ncol = nlat x nlon)
ds_grid = xr.open_dataset(grid_file) 
````

## Reshape & Format Data

````python
# List of ML output variables (mean)
out_vars = ['ptend_t','ptend_q0001','cam_out_NETSW','cam_out_FLWDS','cam_out_PRECSC',
            'cam_out_PRECC','cam_out_SOLS','cam_out_SOLL','cam_out_SOLSD','cam_out_SOLLD']

# Assign length of each output variable
out_vars_len = {'ptend_t'       :60,
                'ptend_q0001'   :60,
                'cam_out_NETSW' :1,
                'cam_out_FLWDS' :1,
                'cam_out_PRECSC':1,
                'cam_out_PRECC' :1,
                'cam_out_SOLS'  :1,
                'cam_out_SOLL'  :1,
                'cam_out_SOLSD' :1,
                'cam_out_SOLLD' :1}

# Set dimemsion names for xr.Datasets
lev_dim_name    = 'lev'
sample_dim_name = 'sample'

# Map the length of dimension to the name of dimension
len_to_dim = {60:lev_dim_name,n_samples:sample_dim_name}
````

````python
# Define number of samples and columns
n_samples = y_pred.shape[0] 
n_ncol    = len(ds_grid['ncol'])      

# Make area-weights
ds_grid['area_wgt'] = ds_grid['area']/ds_grid['area'].mean('ncol')

# Map ds_grid.ncol to n_samples dimension of .npy-loaded arrays (e.g., y_pred)
to_xarray = {'area_wgt':(sample_dim_name,np.tile(ds_grid['area_wgt'], int(n_samples/len(ds_grid['ncol']))))}
to_xarray = xr.Dataset(to_xarray)

# Add n_sample-mapped grid variables back to ds_grid
ds_grid = xr.merge([ds_grid[['P0','hyai','hyam','hybi','hybm','lat','lon','area']],to_xarray[['area_wgt']]])
````

````python
data = {}
for ds in ['true','pred']:
    if ds=='true':
        work = y_true
    elif ds=='pred':
        work = y_pred

    # Construct dictionary for xr.Dataset where key = variable name, value = (dimension names, data)
    to_xarray = {}
    for i,var in enumerate(out_vars):

        # Set dimensions of variable based on number of levels (var_len)
        var_len = out_vars_len[var]
        if var_len==60:
            var_dims = (sample_dim_name,level_dim_name)
        elif var_len == 1:
            var_dims = sample_dim_name

        # Set variable start/end indices in the loaded np.array
        if k==0:
            start_idx = 0
        end_idx = start_idx+var_len
        var_data = np.squeeze(work[:,start_idx:end_idx]) # scaled output
        var_data = var_data/out_scale[var].values # unscaled output
        
        # Add 'var':(var_dims,<np.array>) to dictionary and reset indices
        to_xarray[var] = (var_dims,var_data)
        start_idx = end_idx

    # Convert dictionary to xr.Dataset
    data[ds] = xr.Dataset(to_xarray)

    # Add surface pressure ('state_ps') from ML input
    state_ps = xr.DataArray(x_true[:,120],dims=('sample'),name='state_ps') # normalized ps
    state_ps = state_ps*(in_max['state_ps']-in_min['state_ps'])+in_mean['state_ps'] # denormalized ps
    data[ds]['state_ps'] = state_ps

    # Add grid information
    data[ds] = xr.merge([data[ds],ds_grid])

    # Add pressure thickness of each level (dp), where pressure p(x,z) = hyam(z)*p0 + hybm(z)*ps(x,z)
    tmp = data[ds]['P0']*data[ds]['hyai']+data[ds]['state_ps']*data[ds]['hybi']
    tmp = tmp.isel(ilev=slice(1,61)).values-tmp.isel(ilev=slice(0,60)).values
    tmp = tmp.transpose()
    data[ds]['dp'] = xr.DataArray(tmp,dims=('sample','lev'))

    # Break dimension (sample) into (ncol,time)
    n_timestep   = int(n_samples/n_ncol)
    dim_ncol     = np.arange(n_ncol)
    dim_timestep = np.arange(n_timestep)
    new_idx  = pd.MultiIndex.from_product([dim_timestep,dim_ncol],names=['time','ncol'])
    data[ds] = data[ds].assign_coords(sample=new_idx).unstack('sample')

del work,to_xarray,y_true,y_pred,x_true,state_ps,tmp
````

## Unit Conversion, Area Weighting, & Vertical Weighting

````python
# Define physical constants needed for weighting
g   = 9.80616   
cp  = 1.00464e3
lv  = 2.501e6
lf  = 3.337e5 
ls  = lv+lf
rho_air   = 101325./(6.02214e26*1.38065e-23/28.966)/273.15
rho_water = 1.e3

# Set scaling factors for each variable needed to convert to a common energy unit (W/m2)
out_vars_conv = {'ptend_t'       :cp,
                 'ptend_q0001'   :lv,
                 'cam_out_NETSW' :1.,
                 'cam_out_FLWDS' :1.,
                 'cam_out_PRECSC':lv*rho_water,
                 'cam_out_PRECC' :lv*rho_water,
                 'cam_out_SOLS'  :1.,
                 'cam_out_SOLL'  :1.,
                 'cam_out_SOLSD' :1.,
                 'cam_out_SOLLD' :1.}
````

````python
ds_energy = {}
for ds in ['true','pred']:
    
    # Make a copy to keep original dataset
    ds_energy[ds] = ds[ds].copy(deep=True)

    # vertical weighting / area weighting / unit conversion
    for var in out_vars:
        
        # For 3D variables, weight vertical levels by dp/g (i.e., mass of air within a grid cell per unit area)
        if out_vars_len[var]==60:
            ds_energy[ds][var] = ds_energy[ds][var]*ds_energy[ds]['dp']/g

        # For all variables, weight horizontal area of each grid cell by a[x]/mean(a[x])
        ds_energy[ds][var] = ds_energy[ds]['area_wgt']*ds_energy[ds][var]

        # For variables with different units, convert units to common energy unit (i.e., W/m2)
        ds_energy[ds][var] = out_vars_conv[var]*ds_energy[ds][var]
````

## Calculate & Save Metrics Based on Globally-Averaged Model Output

````python
# Define all metrics
all_metrics = ['MAE','RMSE','R2']

# Subset for output variables
ds_energy[ds] = ds_energy[ds][out_vars]

# Caclulate metrics (3D variables will have dimensions of (ncol,lev) and 2D variables (ncol))
metrics = {}
metrics['MAE']  = (np.abs(ds_energy['true']-ds_energy['pred'])).mean('time')
metrics['RMSE'] = np.sqrt(((ds_energy['true']-ds_energy['pred'])**2.).mean('time'))
metrics['R2']   = 1-((ds_energy['true']-ds_energy['pred'])**2.).sum('time')/\
                    ((ds_energy['true']-ds_energy['true'].mean('time'))**2.).sum('time')

# Take domain mean (3D variables will have dimensions of (lev) and 2D variables will be scalars)
for metric in all_metrics:
    metrics[metric] = metrics[metric].mean('ncol')

# Save domain-mean (3D) results 
metrics_3D = {}
for metric in all_metrics:
    metrics_3D[metric] = metrics[metric].to_stacked_array('ml_out_idx',sample_dims='',name=metric)
work = pd.DataFrame({'MAE' :metrics_3D['MAE'].values,
                     'RMSE':metrics_3D['RMSE'].values,
                     'R2'  :metrics_3D['R2'].values})
work.index.name = 'output_idx'
work.to_csv(metrics_3D_file)

# Save vertically-averaged (2D) results
metrics_2D = {}
for metric in all_metrics:
    metrics_2D[metric] = metrics[metric].mean('lev')
    metrics_2D[metric] = metrics_2D[metric].mean('ilev')
work = pd.DataFrame({'MAE' :metrics_2D['MAE'].to_pandas(),
                     'RMSE':metrics_2D['RMSE'].to_pandas(),
                     'R2'  :metrics_2D['R2'].to_pandas()})
work.index.name = 'Variable'
work.to_csv(metrics_2D_file)
````

# Make Figures

Load in the pre-saved ```.csv``` files (one per model) and plot metrics for model intercomparison.

````python
# File names for saved 2D metrics
metrics_2D_files = {'CNN' :'./metrics/CNN.metrics.lev-avg.csv',
                    'MLP' :'./metrics/MLP.metrics.lev-avg.csv',
                    'RPN' :'./metrics/RPN.metrics.lev-avg.csv',
                    'cVAE':'./metrics/cVAE.metrics.lev-avg.csv',
                    'HSR' :'./metrics/HSR.metrics.lev-avg.csv'}

# File names for saved 3D metrics
metrics_3D_files = {'CNN' :'./metrics/CNN.metrics.csv',
                    'MLP' :'./metrics/MLP.metrics.csv',
                    'RPN' :'./metrics/RPN.metrics.csv',
                    'cVAE':'./metrics/cVAE.metrics.csv',
                    'HSR' :'./metrics/HSR.metrics.csv'}

# Assign model colors
colors = {'CNN' :'#0072B2',
          'HSR' :'#E69F00',
          'MLP' :'#882255',
          'RPN' :'#009E73',
          'cVAE':'#D55E00'}

# Assign variable short names
var_names = {'ptend_t'       :'dT/dt',
             'ptend_q0001'   :'dq/dt',
             'cam_out_NETSW' :'NETSW',
             'cam_out_FLWDS' :'FLWDS',
             'cam_out_PRECSC':'PRECSC',
             'cam_out_PRECC' :'PRECC',
             'cam_out_SOLS'  :'SOLS',
             'cam_out_SOLL'  :'SOLL',
             'cam_out_SOLSD' :'SOLSD',
             'cam_out_SOLLD' :'SOLLD'}

# Assign plotting parameters
abc    = 'abcdefghijk'
sw_log = False
````

### Plot 1: Level-Aggregated Metrics

````python
# Choose models and metrics to plot
models  = ['CNN','HSR','MLP','RPN','cVAE'] 
metrics = ['MAE','RMSE','R2']

# Create dictionary combining all pd.DataFrames of 2D data for each model
data_by_model = {}
for model in models:
    data_by_model[model] = pd.read_csv(metrics_2D_files[model],index_col=0)

# Create dictionary combining all pd.DataFrames for each metric (rows correspond to models)
data_by_metric = {}
for metric in metrics:
    data_by_metric[metric] = pd.DataFrame([data_by_model[model][metric] for model in models],index=models)
````

````python
# Iterate over metrics to plot level-aggregated values for each model
fig,axs = plt.subplots(ncols=len(metrics))
for i,metric in enumerate(metrics):
    ax = axs[i]
    data = data_by_metric[metric]
    data = data.rename(columns=var_short_name)
    data.transpose().plot.bar(color=[colors[model] for model in models],legend=False,ax=ax)
    if metric=='R2':
        ax.set_ylim(0,1)
    ax.set(title=f'({abc[i]}) {metric}',xlabel='',ylabel='$W/m^2$')
    fig.set_size_inches(14,2)
axs[0].legend(loc='upper left',ncols=2,columnspacing=.9,labelspacing=.3,borderpad=.2,
              handleheight=.07,handlelength=1.5,handletextpad=.2)
````

````python
# Iterate over tendency variables and metrics to plot vertical profiles for each model
for var in ['ptend_t','ptend_q0001']:
    fig,axs = plt.subplots(ncols=len(metrics))
    for i,metric in enumerate(metrics):
        ax =axs[i]
        start_idx = var_idx[var][0]
        end_idx   = var_idx[var][1]
        data = data_by_metric[metric].iloc[:,start_idx:end_idx]
        if var=='ptend_q0001':
            data.columns = data.columns-60
            data = data.where(~np.isinf(data),-999)
        data.transpose().plot(color=[colors[model] for model in models],legend=False,ax=ax)
        ax.set(title=f'({abc[i]}) {metric} ({var_names[var]})',xlabel='Level index')
        if k==0:
            ax.set_ylabel('$W/m^2$')
        else:
            ax.set_ylabel('')
        if metric=='R2':
            ax.set_ylim(0,1.05)
        fig.set_size_inches(10,2.5)
    axs[0].legend(loc='upper left',ncols=1,labelspacing=.3,borderpad=.3,
                  handleheight=.07, handlelength=1.5, handletextpad=.2)
````
